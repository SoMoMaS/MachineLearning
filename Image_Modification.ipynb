# Resizing picture

```python
from PIL import Image
import numpy as np
```

```python
width = 96
height = 96

basePath = "./TestData/"
```

```python
import os

def load_images_from_folder(path):
    images = []
    for filename in os.listdir(path):
        img = Image.open(os.path.join(path,filename))
        if img is not None:
            images.append(img)
    return images


images = load_images_from_folder(basePath)
```

```python

images = load_images_from_folder(basePath)


def transform_Pictures(images):
    resized_Pictures = []
    i = 0
    for image in images:
        new_image = image.resize((width, height))
        new_image = new_image.transpose(Image.ROTATE_270)
        new_image.save(basePath + str(i) + 'resized.jpg')
        i += 1
        resized_Pictures.append(new_image)
    return resized_Pictures
```
