# Resizing picture

```python
from PIL import Image
import numpy as np
```

```python
width = 96
height = 96

basePath = "./TestData/"
```

```python
import os

def load_images_from_folder(path):
    images = []
    for filename in os.listdir(path):
        img = Image.open(os.path.join(path,filename))
        if img is not None:
            images.append(img)
    return images


images = load_images_from_folder(basePath)
```

```python

images = load_images_from_folder(basePath)


def transform_Pictures(images):
    resized_Pictures = []
    i = 0
    for image in images:
        new_image = image.resize((width, height))
        new_image = new_image.transpose(Image.ROTATE_90)
        new_image.save(basePath + str(i) + '_resized.jpg')
        i += 1
        resized_Pictures.append(new_image)
    return resized_Pictures
```

# Collecting pixels of picture

```python
def get_average_pixels(image):
    pixels = image.load()
    average_pixels = []

    for x in range(width):
        for y in range(height):
            pixel = pixels[x, y]
            sum = 0
            for val in pixel:
                sum +=val

            average_pixels.append(sum//len(pixel))

    return average_pixels
```

```python
images = load_images_from_folder(basePath)
resizedPictures = transform_Pictures(images)
avarage_pictures = []
counter = 0
for resizedPicture in resizedPictures:
    average_pixels = get_average_pixels(resizedPicture);
    avarage_pictures.append(average_pixels)
    print(average_pixels)
    avarage_pixels_array = np.array(average_pixels, dtype=np.uint8)
    avarage_pixels_array_formatted = np.reshape(avarage_pixels_array, (96, 96))
    avarage_image = Image.fromarray(avarage_pixels_array_formatted)
    avarage_image.save(basePath + str(counter) + '_avarage.png')
    counter += 1
```

# Writing into csv file

```python
import csv  

header = ['ImageId' ,'Image']
rows = []
id = 0
for pixels in avarage_pictures:
    data = ""
    i = 0
    for pixel in pixels:
        data += str(pixel)
        if i != len(pixels) - 1:
             data += " "
        i += 1
    rows.append([id, data])
    print(data)
    id += 1


with open('./Dataset_original/test_4.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    for row in rows:
        writer.writerow(row)
```
